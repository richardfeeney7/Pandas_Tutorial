# Reading Data in Pandas

    *Note that iloc is a ref to the index location
    *loc is used for finding text relations

```python
# Reading Headers
print(df.columns)

# Read Specific Column
print(df.Name) # Print all names 
print("\nPrint All Names:  \n",df["Name"],"\n")    # This will also print all Names
print("\nPrint Frist 5 names: \n",df["Name"][0:5]) # This will print the frist 5 names
print("\nPrint Specific Columns:\n",df[["Name","Type 1","HP"]]) # Print only the columns we want

# Read Each Row
print("\nPrint Row 2:\n",df.iloc[1]) # Print what is in index 1 (This will be row 2)
print("\nPrint Row 0-3:\n",df.iloc[0:3]) # Print what is in index 0 -3

#for index,row in df.iterrows():
    #print("\n For LOOP :\n",index,row)
    
print("\nloc to find Fire in Type 1 : \n", df.loc[df["Type 1"]== "Fire"]) # use loc to find all entries of Fire in Type 1

# Read a specific location (R,C)
print("\nPrint what is in the first position of the second column:\n",df.iloc[2,1])
```

```python
# Next I will all all the stats (HP	Attack	Defense	Sp. Atk	Sp. Def	Speed) to see which pokemon is the best 

df["Total"] = df["HP"]+df["Attack"]+df["Defense"]+df["Sp. Atk"]+df["Sp. Def"]+df["Speed"]
df.head()

# Faster method to this is to use iLoc
df["Total2"] = df.iloc[:,4:10].sum(axis=1) # Axis to 1 mean Hor addition, 0 means vir addition
df.head()
```

# Filtering Data

```python
df.loc[df["Type 1"] == "Grass"] # Only show rows where type 1 matches Grass
```

```python
df.loc[(df["Type 1"] == "Grass") & (df["Type 2"] == "Poison")] # Mach rows from different columns 
```

```python
df.loc[(df["Type 1"] == "Grass") | (df["Type 2"] == "Poison")]
```

```python
df.loc[(df["Type 1"] == "Grass") & (df["Type 2"] == "Poison") & (df["HP"]>70)] # Mach rows from different columns
```

```python
# Can save this to a new dataframe and save to file
new_df = df.loc[(df["Type 1"] == "Grass") & (df["Type 2"] == "Poison") & (df["HP"]>70)] # Mach rows from different columns

new_df.to_csv("filtered.csv", index=False) # Save the new DF

new_df = new_df.reset_index(drop=True) # Reset the index and drop the old index
# new_df.reset_index(drop=True, inplace = True) # We dont need to ceeate a new DF, we can use inplace to update the given 


new_df # Show DF with index reset completed 
```

```python
# Filter all names that contain mega. 
df.loc[df["Name"].str.contains("Mega")].head(20)
```

```python
# drop all names that contain mega. 
df.loc[~df["Name"].str.contains("Mega")] # ~ means not when using loc function
```

```python
# Find Grass or Fire

import re # Import regular expressions

df.loc[df["Type 1"].str.contains("grass|Fire", flags=re.I, regex=True) ].head(20) # Flags is used to ignore case
```

```python
# find everything that starts with pi
df.loc[df["Name"].str.contains("^pi[a-z]*", flags=re.I, regex=True) ].head(20) # Flags is used to ignore case
```
